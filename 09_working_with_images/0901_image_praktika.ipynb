2. Padarykite savo anksčiau sukurtą ikonėlę juodai baltą.

```python
# 1.
from PIL import Image

image = Image.open('image.jpg')
width, height = image.size
crop_area = (850, 330, 980, 460)
cropped_image = image.crop(crop_area)
cropped_image.save('image_cropped.jpg')
image_cropped = Image.open('image_cropped.jpg')
new_size = (128, 128)
resized_image = cropped_image.resize(new_size)
resized_image
# resized_image.show() atvertų naujame lange.
```

```python
# 2.
from PIL import Image
image = Image.open('image_cropped.jpg')
image_black_and_white = image.convert('L')
image_black_and_white.save('image_black_and_white.jpg')
image_black_and_white
```
